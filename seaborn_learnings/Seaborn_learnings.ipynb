```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#seaborn_Settings
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8,4)})
```

# Loading_DataSets

```python
titanic = pd.read_csv('data_for_seaborn_learnings/titanic.csv')
server= pd.read_csv('data_for_seaborn_learnings/serverdata.csv')
mlb = pd.read_csv('data_for_seaborn_learnings/mlbsalaries.csv')
```

# Histograms

```python
plt.hist(titanic.Age.dropna(), bins =25 )
```

```python
plt.hist(server.Time, 25, color=sns.desaturate("indianred",1))
```

```python
plt.hist(mlb.Salary.dropna(), 25)
```

# Box Plots

```python
sns.boxplot(titanic.Age.dropna())
```

```python
sns.boxplot(titanic.Age, titanic.Sex)
```

```python
sns.boxplot(server.Time, orient='v')
```

```python
sns.boxplot(mlb.Salary, orient='v')
```

# kernel Density Estimate Plot

```python
sns.kdeplot(titanic.Age.dropna(), shade=True)
```

```python
sns.kdeplot(server.Time.dropna(), shade=True)
```

```python
sns.kdeplot(mlb.Salary.dropna(), shade=True)
```

```python
sns.distplot(titanic.Age.dropna())
```

```python
sns.distplot(server.Time.dropna())
```

```python
sns.distplot(mlb.Salary.dropna())
```

# Violin Plots

```python
sns.violinplot(titanic.Age.dropna())
```

```python
sns.violinplot(titanic.Age.dropna(), titanic.Sex)
```

```python
sns.violinplot(mlb.Salary)
```

# Cumulative Distribution Function

```python
sns.kdeplot(titanic.Age.dropna(), cumulative=True)
```

```python
sns.kdeplot(server.Time.dropna(), cumulative=True)
```

```python
sns.kdeplot(mlb.Salary.dropna(), cumulative=True)
```
